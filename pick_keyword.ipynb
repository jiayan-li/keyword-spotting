Count words

```python
from collections import Counter
import re
import json
import pickle
```

```python
# read PROMPTS.txt
prompt_file = 'timit/PROMPTS.txt'
with open(prompt_file, 'r') as file:
    prompt = file.readlines()
    # remove the newline character
    prompt = [line.strip() for line in prompt]

prompt
```

```python
# Tokenize and count words
word_counts = Counter()
for sentence in prompt:
    words = re.findall(r'\b\w+\b', sentence)  # Tokenize words
    word_counts.update(words)

# Count types of words
num_types = len(word_counts)
```

```python
word_counts
```

```python
# Save Counter object
with open('word_counts.pickle', 'wb') as f:
    pickle.dump(word_counts, f)
```

```python
# Load Counter object
with open('word_counts.pickle', 'rb') as f:
    loaded_word_counts = pickle.load(f)
```
